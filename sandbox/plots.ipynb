```python
import numpy as np
import matplotlib.pyplot as plt

data_sizes = np.array([1e3, 1e4, 1e5, 1e6, 1e7], dtype=int)

mean_a = np.array([0.25, 0.20, 0.18, 0.15, 0.14])
std_a  = np.array([0.05, 0.04, 0.03, 0.025, 0.02])

mean_b = np.array([0.30, 0.22, 0.17, 0.13, 0.11])
std_b  = np.array([0.06, 0.05, 0.04, 0.03,  0.025])

mean_c = np.array([0.22, 0.19, 0.16, 0.155, 0.145])
std_c  = np.array([0.07, 0.055, 0.045, 0.035, 0.03])

fig, ax = plt.subplots()

ax.plot(data_sizes, mean_a, marker='o', color='steelblue', label='Model A mean')
ax.fill_between(data_sizes, mean_a - std_a, mean_a + std_a,
                alpha=0.2, color='steelblue', label='Model A ±1 std dev')

ax.plot(data_sizes, mean_b, marker='s', color='coral', label='Model B mean')
ax.fill_between(data_sizes, mean_b - std_b, mean_b + std_b,
                alpha=0.2, color='coral', label='Model B ±1 std dev')

ax.set_xlabel('Dataset Size')
ax.set_ylabel('Average Error')
ax.set_title('Error vs Dataset Size with Standard Deviation')
ax.set_xscale('log')
ax.legend()
ax.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data_sizes = np.array([1e3, 1e4, 1e5, 1e6, 1e7], dtype=int)
labels = ['1K', '10K', '100K', '1M', '10M']

mean_a = np.array([0.25, 0.20, 0.18, 0.15, 0.14])
std_a  = np.array([0.05, 0.04, 0.03, 0.025, 0.02])

mean_b = np.array([0.30, 0.22, 0.17, 0.13, 0.11])
std_b  = np.array([0.06, 0.05, 0.04, 0.03,  0.025])

mean_c = np.array([0.22, 0.19, 0.16, 0.155, 0.145])
std_c  = np.array([0.07, 0.055, 0.045, 0.035, 0.03])

x = np.arange(len(labels))
width = 0.25

fig, ax = plt.subplots(figsize=(9, 5))

ax.bar(x - width, mean_a, width, yerr=std_a, label='Model A',
       color='steelblue', alpha=0.75, capsize=4, error_kw={'linewidth': 1.5})
ax.bar(x,         mean_b, width, yerr=std_b, label='Model B',
       color='coral', alpha=0.75, capsize=4, error_kw={'linewidth': 1.5})
ax.bar(x + width, mean_c, width, yerr=std_c, label='Model C',
       color='mediumseagreen', alpha=0.75, capsize=4, error_kw={'linewidth': 1.5})

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_xlabel('Dataset Size')
ax.set_ylabel('Average Error')
ax.set_title('Error vs Dataset Size with Error Bars')
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data (replace with your own)
data_sizes = np.array([1e3, 1e4, 1e5, 1e6, 1e7], dtype=int)
mean_error = np.array([0.25, 0.20, 0.18, 0.15, 0.14])
std_dev = np.array([0.05, 0.04, 0.03, 0.025, 0.02])

# Compute bounds
upper = mean_error + std_dev
lower = mean_error - std_dev

# Plot
plt.figure()
plt.plot(data_sizes, mean_error, marker='o', label='Mean Error')
plt.fill_between(data_sizes, lower, upper, alpha=0.1, label='±1 Std Dev')

# Labels and title
plt.xlabel('Dataset Size')
plt.ylabel('Average Error')
plt.title('Error vs Dataset Size with Standard Deviation')
plt.legend()
plt.grid()
plt.xscale('log')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data (replace with your own)
data_sizes = np.array([1e3, 1e4, 1e5, 1e6, 1e7], dtype=int)
mean_error = np.array([0.25, 0.20, 0.18, 0.15, 0.14])
std_dev = np.array([0.05, 0.05, 0.05, 0.05, 0.07])

upper = mean_error + std_dev
lower = mean_error - std_dev

# Plot
plt.figure()
plt.errorbar(
    data_sizes,
    mean_error,
    yerr=std_dev,
    fmt='o',          # markers + line
    capsize=5,         # little caps on error bars
    label='Mean Error ± Std Dev'
)

plt.fill_between(data_sizes, lower, upper, alpha=0.1, label='±1 Std Dev')


# Labels and title
plt.xlabel('Dataset Size')
plt.ylabel('Average Error')
plt.title('Error vs Dataset Size (Error Bars)')
plt.legend()
plt.grid()
plt.xscale('log')

plt.show()
```
